```python
#now changing some of the data set
'''
This is a Python function called "parse" that takes a string argument "x".

Inside the function, the string "x" is split into a list of substrings using the split() method, 
and the second substring (index 1) is sliced to only include the first 8 characters.

Then, a new string "z" is created by concatenating the first substring (index 0) and 
the sliced second substring with a hyphen in between.

Finally, the "datetime" module is used to convert the string "z" into a datetime object "d" with a 
specific format of 'YYYY-MM-DD HH:MM:SS'.
'''
def parse(x):
    y = x.split()
    t = y[1][:8]
    z=y[0] + " " + t
    d = datetime.strptime(z, '%Y-%m-%d %H:%M:%S')
    return d
```

```python
df = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/data-science-in-health-care-basic-statistical-analysis/COVID_19.xlsx',
    'Sheet1',
    na_values="NaN",
    parse_dates=['Date time'],
    index_col=0,
    date_parser=parse)
```

```python
df
```

```python
# if we want to remove the rows that do not provide proper data then we can use dropna() method
# in our case some respondents didn't indicate their gender 
df = df.dropna(subset=['Gender'])
```

```python
'''Some data might be stored in formats that Python misinterprets. Therefore, you can use the map() method to replace cell values.
 For example, you can change the Yes / No value to True / False as follows:'''
d = {'No' : False, 'Yes' : True}
c = 'Do you vaccinated influenza?'
df.loc[:, c] = df[c].map(d)
```

```python
df
```

```python
#For further analysis it is necessary to study the general characteristics of the fields using info()
df.info()
```

```python
c = 'Age'
df.loc[:, c] = df[c].astype('category')
```

```python
df['Age']
```

```python
for c in df.columns[1:-1]:
    df.loc[:, c] = df[c].apply(lambda x : str(x) if str(x).find('(') == -1 else str(x)[:str(x).find('(')] ).astype('category')
```

```python
# Now let's visualize our data. We will use SeaBorn library.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualizing the number of surveyed men and women in terms of age groups using countplot():
sns.countplot(x='Age', hue='Gender', data=df)
```

```python
# Let's analyze the dynamics of surveys during the whole time of research. 
# To do this, you should group the data by day using the resample() and plot() methods.
df.resample(str('1D')).sum().plot()
```

```python
# Some of the surveyed patients were sure they had COVID-19 and others weren't. 
# Let's plot the temperature distribution (distplot()) of both groups on one figure that contains two graphs (subplots())
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))

df_t = df[df['Have you had Covid`19 this year?'] == 'Yes'].dropna(subset=['Maximum body temperature'])
sns.distplot(df_t['Maximum body temperature'], ax=axes[0])
df_t = df[df['Have you had Covid`19 this year?'] == 'Maybe'].dropna(subset=['Maximum body temperature'])
sns.distplot(df_t['Maximum body temperature'], ax=axes[1])
```

```python
# Let's build temperature dependence on age group (boxplot() or violinplot()).
#  The graph shows the data quarters. The midline of the rectangle is the median. The whiskers show the data scatter.
cols = ['Maximum body temperature', 'Maximum body temperature']
_, axes = plt.subplots(1, 2, sharey=True, figsize=(16,6))
sns.boxplot(y = df["Age"], x=df["Maximum body temperature"], ax=axes[0])
sns.violinplot(y = df["Age"], x=df["Maximum body temperature"],ax=axes[1] )
```
